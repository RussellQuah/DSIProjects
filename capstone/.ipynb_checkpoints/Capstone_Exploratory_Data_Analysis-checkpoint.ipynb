```python
# importing libraries
import numpy as np
import random
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Importing visualisation libraries
import seaborn as sns
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline

# NLP libraries
from bs4 import BeautifulSoup
import re
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
```

## Data Cleaning
An annual report released on 2020 would be the annual report for the year ending 2019.

```python
# dropping all the missing values.
df.dropna(inplace=True)
```

```python
# calculating the percentage price change
df['price_change_%'] = df['price_change_next_yr']/df['price_on_date']
df['sp_price_change_%'] = df['sp_price_change_next_year']/df['sp_on_date']
```

```python
#further cleaning of xml
df['reports'] = df['reports'].map(lambda x: str(x).replace('\x97', '').replace('\u200b', '').replace('\x94', '').replace('\x92', ''))
```

```python
# Addiotional Stopwords that are relevant to financial statements
stop_words = ['the', 'Company', 'fair', 'value', 'financial', 'statement']
```

```python
df.shape
```

```python
df.head()
```

## Pre-Processing

```python
# Instantiating Tokenizer, PorterStemmer and lemmatizer

tokenizer = RegexpTokenizer(r'\w+')
p_stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
#taken from DSI-14 week 5.03 NLP startercode
def pre_processing(raw_text):

    # 1. Remove HTML.
    review_text = BeautifulSoup(raw_text).get_text()
    
    # 2. Remove non-letters.
    letters_only = re.sub("[^a-zA-Z]", " ", review_text)
    
    # 3. Convert to lower case, split into individual words.
    words = letters_only.lower().split()
    
    # 4. Searching a set is much faster than searching a list
    #    Adding in additional stopwords.
    stops = set(stopwords.words('english') + stop_words)
    
    # 5. Remove stopwords.
    meaningful_words = [w for w in words if w not in stops]
    
    
    #5.5 lemmatizing of words
    meaningful_words = [lemmatizer.lemmatize(w) for w in words]
    
    # 6. Join the words back into one string separated by space, 
    # and return the result.
    return(" ".join(meaningful_words))
```

```python
# Get the number of posts on the dataframe size.
total_posts= df.shape[0]
print(f'There are {total_posts} posts')
```

```python
# Initialize an empty list to hold the clean posts.
clean_reports = []

print("Cleaning and parsing the annual reports...")

# Instantiate counter.
j = 0

# For every post in our training set...
for train_posts in df['reports']:
    
    # Convert post to words, then append to clean_train_posts.
    clean_reports.append(pre_processing(train_posts))
    
    # If the index is divisible by 1000, print a message.
    if (j + 1) % 100 == 0:
        print(f'Post {j + 1} of {total_posts}.')
    
    j += 1
```

```python
df= df.drop(['reports'],axis=1).join(pd.DataFrame(clean_reports,index=df.index, columns=['clean_reports']))
```

```python
df.head()
```

## Target labelling

```python
target = []

for price, sp in zip(df['price_change_%'], df['sp_price_change_%']):
    if price > sp:
        target.append(1)
    elif price <= sp:
        target.append(0)

df['target'] = target        
```
