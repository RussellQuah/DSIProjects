```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train = pd.read_csv('./datasets/predict-west-nile-virus/train.csv/train.csv')
df_test = pd.read_csv('./datasets/predict-west-nile-virus/test.csv/test.csv')
spray = pd.read_csv('./datasets/predict-west-nile-virus/spray.csv/spray.csv')
weather = pd.read_csv('./datasets/predict-west-nile-virus/weather.csv/weather.csv')
```

```python
# Experiment space
# Make a copy of train
train_copy = df_train.copy()
# sort by date, trap, species, in ascending order (specificied in list order)
train_copy.sort_values(['Date', 'Trap','Species'], ascending=[True, True,True],inplace=True)
# Create separate dataframe that sums up the mosquitos with same date, trap, and species
train_ported = train_copy[0:1]
# Use set_value() to cumulate sum mosquito
# Syntax:DataFrame.set_value(index, col, value, takeable=False).
# For col, Use .get_loc() to return integer location, slice or boolean mask for requested label."""
#col_mosquito = train_copy.columns.get_loc("NumMosquitos")
# Loop through copy of train from the second row
# Compare and Sum(on train_ported) the mosquitos 
# if date, trap, and species same(current row & previous row) across dataframes
for i in range(1,len(train_copy)):
    if ((train_copy.iloc[i]['Date'] == train_ported.iloc[-1]['Date'])&\
        (train_copy.iloc[i]['Trap'] == train_ported.iloc[-1]['Trap'])&\
        (train_copy.iloc[i]['Species'] == train_ported.iloc[-1]['Species'])):
        #train_ported.set_value(-1,col_mosquito, (train_ported.iloc[-1]["NumMosquitos"]+train_copy.iloc[i]["NumMosquitos"]),takeable=True)
        train_ported["NumMosquitos"].at[i-1] = train_ported.iloc[-1]["NumMosquitos"]+train_copy.iloc[i]["NumMosquitos"]
    else:
        train_ported = train_ported.append(train_copy[i:i+1])
```
